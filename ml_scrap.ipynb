```python
import pandas as pd
import numpy as np

# Initialize an empty list to store the combined X and Y values
X_combined = []

# Loop through each file and read X and Y values
for i in range(len(nion)):
    df = pd.read_csv(files[i], sep="\t", header=None, 
                     names=["z", "xHI"])
    z = df["z"].values.astype(float)  # Convert to numpy array
    xH = df["xHI"].values.astype(float)  # Convert to numpy array
    
    # Ensure that both z and xH have the same length
    min_length = min(len(z), len(xH))
    z = z[:min_length]
    xH = xH[:min_length]

    # Combine X and Y values for each nion and append to X_combined
    combined_values = np.column_stack((z, xH))
    X_combined.extend(combined_values)

# Convert the list of arrays into a single numpy array
X_combined = np.array(X_combined)

# Now X_combined contains the combined X and Y values for each nion
```

```python
import pandas as pd
import numpy as np

# Initialize an empty list to store combined X and Y values for each nion
X_combined_per_nion = []

# Loop through each file and read X and Y values
for i in range(len(nion)):
    df = pd.read_csv(files[i], sep="\t", header=None, 
                     names=["z", "xHI"])
    z = df["z"].values.astype(float)  # Convert to numpy array
    xH = df["xHI"].values.astype(float)  # Convert to numpy array
    
    # Ensure that both z and xH have the same length
    min_length = min(len(z), len(xH))
    z = z[:min_length]
    xH = xH[:min_length]

    # Combine X and Y values for each nion and append to X_combined_per_nion
    combined_values = z+xH
    X_combined_per_nion.append(combined_values)

# Now X_combined_per_nion contains the combined X and Y values for each nion
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
 
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)
 
# define input sequence
# choose a number of time steps
# split into samples
#X, y = split_sequence(raw_seq, n_steps)
X=reshaped_array
y=nion
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
# define model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(len(reshaped_array[0]), n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=50, verbose=1)

```
